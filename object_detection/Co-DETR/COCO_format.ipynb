```python
def clean_text(text):
    # Replace underscore with a space
    text = text.replace('_', ' ')
    # Remove anything within parentheses (including the parentheses)
    text = re.sub(r'\(.*?\)', '', text)
    return text.strip().lower()
```

```python
cat_ids = [i for i in range(1, len(CLASSES)+1)]
cat2label = {cat_id: i for i, cat_id in enumerate(cat_ids)}
```

```python
import re
```

```python
result_with_lvis_label  = []
for output in outputs:
    item = {}
    for idx in range(len(output)):
        text_label = CLASSES[idx]
        text_label = clean_text(text_label)
        text_label = text_label.lower()
        if text_label in coco_symset.keys():
            text_label = coco_symset[text_label]
        item[text_label] = output[idx]
    result_with_lvis_label.append(item)
        
```

```python
lvis_coco_maping = {}
with open('/home/ngoc/githubs/aux/object_detection/labels_setup/coco_labels.txt', 'r') as file:
    for line in file:
        lvis_coco_maping[line.strip().lower()] = len(lvis_coco_maping)
```

```python
result_with_coco_idx = []
for result in result_with_lvis_label:
    i = 0
    lvis_labels  = result.keys()
    item  = {}
    for lvis_label in lvis_labels:   
        if lvis_label in lvis_coco_maping:
            coco_index = lvis_coco_maping[lvis_label]
            item[coco_index] = result[lvis_label]
    result_with_coco_idx.append(item)
        
```

```python
coco_result = []
```

```python
empty_array = np.empty((0, 5), dtype=np.float32)
```

```python
for result in result_with_coco_idx:
    item  = [result.get(key, empty_array) for key in range(80)]
    coco_result.append(item)
```
